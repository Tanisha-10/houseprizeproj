In scikit-learn, most transformers expect a 2D array-like object as input. When you use housing["population"], you're selecting a single column from the DataFrame housing, which returns a pandas Series. However, scikit-learn transformers typically expect a 2D array-like object, where the first dimension represents the samples and the second dimension represents the features.

By using housing[["population"]], you're selecting a DataFrame with a single column, resulting in a 2D array-like object (specifically, a DataFrame). This matches the expected input format of scikit-learn transformers.

You could alternatively use housing["population"].values.reshape(-1, 1) to convert the pandas Series to a 2D numpy array explicitly, but using housing[["population"]] achieves the same result in a more concise manner.

```python
rbf_transformer = FunctionTransformer(rbf_kernel , kw_args=dict(Y=[[35.]] ,
                                                                gamma=0.1))
age_simil_35 = rbf_transformer.transform(df[["housing_median_age"]])
#radial basis function kernel ....
```

RADIAL basis function kernel calculates the similarity between two points based on their Euclidean distance in a high-dimensional space.

RBF kernel is commonly used because-

**Flexibility:** The RBF kernel is a non-linear kernel, it can model complex relationships between data points and can capture intricate patterns that linear models cannot.

**Universal Approximation**: The RBF kernel has the property of universal approximation, it can approximate any continuous function to arbitrary accuracy given enough hidden units and an appropriate choice of parameters. makes it very powerful for modeling a wide range of functions and data distributions.

Implicit Feature Mapping: The RBF kernel implicitly maps the data into a higher-dimensional space where the data might be linearly separable.

Smoothness: The RBF kernel produces smooth decision boundaries, which can lead to more stable and generalizable models.

Efficiency: Despite its ability to model complex relationships, the RBF kernel can be computationally efficient, especially with modern optimization techniques and hardware acceleration

```python
ratio_transformer = FunctionTransformer(lambda X: X[:, [0]] /X[:, [1]])
#colon comma means it is selecting all rows
```

X[[0]] selects a single row from the array X.

X[:, [0]] selects a single column from the array X.

: selects all rows of X, while [0] specifies that only the first column of X should be selected.

In Python, when you use X[:][1], it first selects all rows of the 2D array X (equivalent to X[:]), and then it attempts to select the second element from the resulting array. However, this won't give you the desired result of selecting the second row of the array. Instead, it will select the second element of the first row due to the order of operations in Python.

X[:,[1]]  what will it give?
will return a 2D array with all rows and only the values from the second column of X.it selects it as a list

X[:,1]  what will it give?
will return a 1D array containing the values from the second column of X.it selects directly

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_array, check_is_fitted
```

```python
class StandardScalerClone(BaseEstimator, TransformerMixin):
 def __init__(self, with_mean=True): # no *args or **kwargs!
   self.with_mean = with_mean
 def fit(self, X, y=None): # y is required even though we don't use it
   X = check_array(X) # checks that X is an array with finite float values
   self.mean_ = X.mean(axis=0)
   self.scale_ = X.std(axis=0)
   self.n_features_in_ = X.shape[1] # every estimator stores this in fit()
   return self # always return self!
 def transform(self, X):
   check_is_fitted(self) # looks for learned attributes (with trailing _)
   X = check_array(X)
   assert self.n_features_in_ == X.shape[1]
   if self.with_mean:
      X = X - self.mean_
   return X / self.scale_
```

```python
from sklearn.cluster import KMeans

class ClusterSimilarity(BaseEstimator, TransformerMixin):
  def __init__(self, n_clusters=10, gamma=1.0,random_state=None):
   self.n_clusters = n_clusters
   self.gamma = gamma
   self.random_state = random_state
  def fit(self, X, y=None, sample_weight=None):
   self.kmeans_ = KMeans(self.n_clusters,random_state=self.random_state)
   self.kmeans_.fit(X, sample_weight=sample_weight)
   return self # always return self!
  def transform(self, X):
   return rbf_kernel(X, self.kmeans_.cluster_centers_,gamma=self.gamma)
  def get_feature_names_out(self, names=None):
   return [f"Cluster {i} similarity" for i in
range(self.n_clusters)]
```

Preprocessing pipeline takes the entire training dataset and applies each transformer to the appropriate columns, then concatenates the transformed columns horizontally (transformers must
never change the number of rows).

```python
def column_ratio(X):
 return X[:, [0]] / X[:, [1]]
def ratio_name(function_transformer, feature_names_in):
 return ["ratio"] # feature names out
def ratio_pipeline():
 return make_pipeline(
 SimpleImputer(strategy="median"),
 FunctionTransformer(column_ratio,
feature_names_out=ratio_name),
 StandardScaler())
log_pipeline = make_pipeline(
 SimpleImputer(strategy="median"),
 FunctionTransformer(np.log, feature_names_out="one-to-one"),
 StandardScaler())
cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1.,
random_state=42)
default_num_pipeline = make_pipeline(SimpleImputer(strategy="median"),
 StandardScaler())
preprocessing = ColumnTransformer([
 ("bedrooms", ratio_pipeline(), ["total_bedrooms",
"total_rooms"]),
 ("rooms_per_house", ratio_pipeline(), ["total_rooms",
"households"]),
 ("people_per_house", ratio_pipeline(), ["population",
"households"]),
 ("log", log_pipeline, ["total_bedrooms", "total_rooms",
"population",
 "households", "median_income"]),
 ("geo", cluster_simil, ["latitude", "longitude"]),
 ("cat", cat_pipeline,
make_column_selector(dtype_include=object)),
 ],
 remainder=default_num_pipeline) # one column remaining : housing_median_age
```
